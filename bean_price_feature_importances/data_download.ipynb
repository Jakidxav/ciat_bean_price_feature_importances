### Loading price data

```python
cameroon_drc_kenya_data_path = './data/raw/price_cameroon_drc_kenya.csv'
rwanda_uganda_tanzania_data_path = './data/raw/price_rwanda_uganda_tanzania.csv'
```

```python
cameroon_drc_kenya = pd.read_csv(cameroon_drc_kenya_data_path)
rwanda_uganda_tanzania = pd.read_csv(rwanda_uganda_tanzania_data_path)
```

```python
fig, ax = plt.subplots(figsize=(12, 7))
cameroon_drc_kenya.plot(x='Date-Monthly', ax=ax)
plt.legend(bbox_to_anchor=(1, 1))
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 7))
rwanda_uganda_tanzania.plot(x='Date-Monthly', ax=ax)
plt.legend(bbox_to_anchor=(1, 1))
plt.show()
```
